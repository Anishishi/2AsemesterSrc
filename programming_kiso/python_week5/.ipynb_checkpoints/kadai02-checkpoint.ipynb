nestlist関数の実装
例えば
nestlist(f, x0, 3)
の場合は関数をf,初期値をx0,適用する回数を3回として出力は
[x0, f(x0), f(f(x0)), f(f(f(x0)))]
となる。

```python
def nestlist(func, x0, num):
    l = [x0]
    for i in range(num):
        l.append(func(l[i]))
    return l
```

```python
nestlist(lambda x: x**2, 2, 5)
```

```python
##ランダムウォーク
import random
import matplotlib.pyplot as plt
l = nestlist(lambda x: x+(-1)**random.randint(0,1),0,1000)
plt.plot(l)
plt.show()
```

```python
##Tinkerbell写像
def nestlist(func, x0, num):
    l = [x0]
    for i in range(num):
        l.append(func(*l[i]))
    return l
import random
import matplotlib.pyplot as plt
g = lambda x,y:(x**2-y**2+0.9*x-0.6013*y,2*x*y+2.0*x+0.50*y)
l = nestlist(g ,(-0.72, -0.64), 100000)
x = []
y = []
for i in range(100000):
    x.append(l[i][0])
    y.append(l[i][1])
plt.scatter(x, y, s=0.1)
plt.show()
```
