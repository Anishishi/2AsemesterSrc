一階の微分方程式を数値的に解く。
オイラー法による、多次元に対応したもの。
適当にモジュール化する

```python
def ndsolve(func, x0, t_0, t_last):
    delta_t = 0.01
    repeat_num = int((t_last - t_0)/delta_t)
    t = [t_0]
    x = [x0]
    for i in range(repeat_num):
        temp = []
        for j in range(len(func)):
            temp.append(x[i][j]+delta_t*(func[j](*x[i])))
        t_0 += delta_t
        x.append(temp)
        t.append(t_0)
    ans = []
    for i in range(len(t)):
        temp = []
        temp.append(t[i])
        for j in range(len(func)):
            temp.append(x[i][j])
        ans.append(temp)
    return ans
```

```python
fs = [lambda x,y: x - x**3 - y, lambda x,y: x ]
traj = ndsolve(fs, [2,1], 0, 30)
import matplotlib.pyplot as plt
ttr, xtr, ytr = zip(*traj)
plt.plot(xtr, ytr)
plt.show()
```
