ランダムウォークにおいて、輸送速度をシュミレーションで計算するcプログラムをpython側から呼び出して、解析解と比較する。

```python
import subprocess
import matplotlib.pyplot as plt
import numpy as np
n = 10
ans = []
q_l = []
for i in range(n+1):
    q = (0.9-0.5)/10*i+0.5
    q_l.append(q)
    output = subprocess.run(
        args = ['./kadai4',f'{q}'] , stdout = subprocess.PIPE)
    ans.append(output.stdout.decode())
ans = list(map(float, ans))
plt.scatter(q_l, ans, color='red')
x = np.arange(0.4,1.0,0.01)
y=x/(x+1)
plt.plot(x, y, color='blue')
plt.xlabel("q")
plt.ylabel("velocity")
plt.show()
```
